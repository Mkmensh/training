# GitHub Training Repository Setup

This notebook demonstrates how to initialize a local Git repository, commit the current project, and push it to `https://github.com/Mkmensh/training.git`. It also includes repository inspection and commit steps.

## 1. Clone the GitHub Repository

If you need to start from a fresh clone, use the repository URL below. For this project, we are initializing the existing local directory instead.

```bash
git clone https://github.com/Mkmensh/training.git
```

````python
import os
print('Working directory:', os.getcwd())
!git init
````

## 2. Inspect Repository Contents

Review the current branch and repository status before making changes.

````python
!git status --short --branch
!git branch --show-current
````

## 3. Create or Modify Files

Add new files and ensure `.gitignore` is configured to ignore build artifacts and IDE metadata.

````python
from pathlib import Path
root = Path('.')
gitignore = root / '.gitignore'
if not gitignore.exists():
    gitignore.write_text(
        '# Maven target directory\n'
        'target/\n\n'
        '# IntelliJ IDEA\n'
        '*.iml\n'
        '.idea/\n\n'
        '# Eclipse files\n'
        '.project\n'
        '.classpath\n'
        '.settings/\n\n'
        '# OS files\n'
        '.DS_Store\n'
        'Thumbs.db\n',
        encoding='utf-8'
    )
    print('.gitignore created')
else:
    print('.gitignore already exists')

!git status --short --branch
````

## 4. Stage and Commit Changes

Stage all changes and create an initial commit message.

````python
!git add .
!git commit -m "Initial commit"
````

## 5. Push Changes to Remote

Add the remote repository and push the local `main` branch to GitHub.

````python
!git branch -M main
!git remote add origin https://github.com/Mkmensh/training.git
!git push -u origin main
````

````python
import os
import subprocess
print('cwd:', os.getcwd())
result = subprocess.run(['git','--version'], capture_output=True, text=True)
print('git version stdout:', result.stdout)
print('git version stderr:', result.stderr)
result = subprocess.run(['git','init'], capture_output=True, text=True)
print('git init stdout:', result.stdout)
print('git init stderr:', result.stderr)
print('returncode:', result.returncode)
````
